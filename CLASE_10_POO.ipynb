```python
class Casa:
  """
  Construir el constructor
  argumentos:
     color(str): color de la casa
     habitaciones(int): el nmero de habitaciones de la casa
     baos(int): el nmero de baos en la casa
     estacionamiento(bool): si tiene o no
     tiene_patio(bool): si tiene patio o no
  """
  def __init__(self, color, numero_habitaciones, numero_banos, estacionamiento, tiene_patio):
    self.color = color
    self.numero_habitaciones = numero_habitaciones
    self.numero_banos = numero_banos
    self.estacionamiento = estacionamiento
    self.tiene_patio = tiene_patio
  def descripcion(self):
    #Describir la casa
    descripcion=f"La casa es de color {self.color}, tiene {self.numero_habitaciones} habitaciones y {self.numero_banos} baos"
    if self.estacionamiento:
      descripcion += " y tiene estacionamiento"
    if self.tiene_patio:
      descripcion += " y tiene patio"
    return descripcion

  def pintar(self,nuevo_color):
    self.color=nuevo_color
    print(f"La casa ha sido pintada de color {self.color}")
```

```python
mi_casa=Casa("azul",3,2,True,True)
type(mi_casa)
```

```python
print(mi_casa.descripcion())
```

```python
mi_casa.numero_banos=5
print(mi_casa.descripcion())
```

```python
mi_casa.pintar("blanco")
print(mi_casa.descripcion())
```

```python
class Floristeria():
  """
  Constructor de la Florsitera
  Argumentos:
     Diseo(str): La casa tiene un diseo floral, pintado en las paredes exteriores
     habitaciones(int): Tiene un gran saln como habitacin
     baos(int): tiene un bao
     tipos_flores(list):Lista de flroes
     nombre_floristeria: La floristera tiene como nombre ROSEMARY
  """
  def __init__(self, diseo, habitaciones, baos, nombre_floristeria, tipos_flores):
    self.diseo = diseo
    self.habitaciones = habitaciones
    self.baos = baos
    self.tipos_flores = tipos_flores
    self.nombre_floristeria=nombre_floristeria
  def descripcion(self):
    #Describir la floristera
    descripcion=f"La floristera se llama {self.nombre_floristeria}. La floristeria {self.diseo}, tiene {self.habitaciones} saln y {self.baos} bao. Estn ofrecidas las siguientes flores: {self.tipos_flores}"
    return descripcion
```
