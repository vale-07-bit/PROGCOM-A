13. lambda: es una función anónima, acorta el proceso.
```
cuadrado = lambda x: x**2
print(cuadrado(4))
```
14. Global y nonlocal: acceso a variables de ámbito superior.
```
x = 10
def f():
    global x
    x = 20
f()
print(x)
```
```
def externo():
    x = 5
    def interno():
        nonlocal x
        x = 10
    interno()
    print(x)  # 10
externo()
```
